```python
def is_same_len(cdr3_1, cdr3_2):
  result = False
  if cdr3_1 == cdr3_2:
    result = True
  return result
```

```python
# (i) 
# they have same length -- check

# (ii) 
# the Hamming distance between their nucleotide sequences is below 10% -- mismatches < 0.1(length of cdr3)

# (iii) 
# they represent VDJ sequences with identical V and J genes. 
# Ignore gene alleles while checking the third condition: 
# e.g., alleles IGHV1-69*01 and IGHV1-69*02 represent the same gene.
```

```python
from tqdm import tqdm
import tensorflow as tf

device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
  print('GPU device not found')
print('Found GPU at: {}'.format(device_name))
```

```python
copy = cdr3_.copy()
# copy['cdr3'][0].replace('G', '1').replace('C', '2').replace('A', '3').replace('T', '4')
```

```python
copy.head()
```

```python
copy.shape
```

```python
hamming_v = []
hamming_j = []
cdr3_len = []
cdr3s = []
# mismatch_len = []

with tf.device('/device:GPU:0'):
  for i in range(copy['cdr3'].shape[0]):
    for j in range(i, copy['cdr3'].shape[0]):
      mismatch = []
      v_str = copy['v_call'][i] == copy['v_call'][j]
      j_str = copy['j_call'][i] == copy['j_call'][j]
      same_len = is_same_len(copy['cdr3_len'][i], copy['cdr3_len'][j])
      if v_str and j_str and same_len:
        for c in range(copy['cdr3_len'][i]):# length of string
          if copy['cdr3'][i][c] != copy['cdr3'][j][c] and len(mismatch) < 0.1*copy['cdr3_len'][i]:
            mismatch.append(c)
            if c == copy['cdr3_len'][i]-1:
              hamming_v.append(copy['v_call'][i])
              hamming_j.append(copy['j_call'][j])
              cdr3s.append(copy['cdr3'][i])
              cdr3_len.append(copy['cdr3_len'][i])
              # mismatch_len.append(len(mismatch))
          elif len(mismatch) > 0.1*copy['cdr3_len'][i]:
            break

# 00:36:24 mins
```
