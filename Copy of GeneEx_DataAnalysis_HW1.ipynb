```r
layout(matrix(c(1,1,1,1,1,1,1,1,2,2), 5, 2, byrow = TRUE))
par(oma=c(5,7,1,1))
cx <- rev(colorpanel(25,"yellow","black","blue"))
leg <- seq(min(dat_cor,na.rm=T),max(dat_cor,na.rm=T),length=10)
options(repr.plot.width=20, repr.plot.height=10)
image(dat_cor,main="Correlation plot Normal/Tumor data",axes=F,col=cx)
axis(1,at=seq(0,1,length=ncol(dat_cor)),label=dimnames(dat_cor)[[2]],cex.axis=2,las=2)
axis(2,at=seq(0,1,length=ncol(dat_cor)),label=dimnames(dat_cor)[[2]],cex.axis=1.5,las=2)
```

```r
dat <- t(t_ren)			#transpose dat
dat.dist <- dist(dat,method="euclidean")	# calculate distance
dat.clust <- hclust(dat.dist,method="single")	# calculate clusters
options(repr.plot.width=15, repr.plot.height=15)
plot(dat.clust,labels=names(dat),cex=1.7)	# plot cluster tree
```

```r
t_ren[t_ren < 0] = NA
df1 = na.omit(t_ren)
```

```r
# cv vs. mean plot
dat.mean <- apply(log2(df1),2,mean)		# calculate mean for each sample
dat.sd <- sqrt(apply(log2(df1),2,var))		# calculate st.deviation for each sample
dat.cv <- dat.sd/dat.mean			#calculate cv
options(repr.plot.width=15, repr.plot.height=15)
plot(dat.mean,dat.cv,main="Alon colon cancer dataset\nSample CV vs. Mean",xlab="Mean",ylab="CV",col='blue',cex=1.5,type="n")
points(dat.mean,dat.cv,bg="lightblue",col=1,pch=21)
text(dat.mean,dat.cv,label=dimnames(df1)[[2]],pos=1,cex=1.5)
```

```r
# average correlation plot
dat.avg <- apply(dat_cor,1,mean)
par(oma=c(3,0.1,0.1,0.1))
options(repr.plot.width=15, repr.plot.height=15)
plot(c(1,length(dat.avg)),range(dat.avg),type="n",xlab="",ylab="Avg r",main="Avg correlation of Tumor/Normal samples",axes=F)
points(dat.avg,bg="red",col=1,pch=21,cex=1.75)
axis(1,at=c(1:length(dat.avg)),labels=dimnames(t_ren)[[2]],las=2,cex.lab=1.4,cex.axis=1.2)
axis(2)
abline(v=seq(0.5,62.5,1),col="grey")
```
