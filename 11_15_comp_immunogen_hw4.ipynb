```python
def as_numeric(gene_sample):
  gene_new = ''
  for i in gene_sample:
    if i == 'A' or i == 'a':
      gene_new += '1'
    if i == 'T' or i == 't':
      gene_new += '2'
    if i == 'G' or i == 'g':
      gene_new += '3'
    if i == 'C' or i == 'c':
      gene_new += '4'
  return int(gene_new)
```

```python
def process(_ref, _template):
  diffs = []
  start_idxs = []
  sum_of_diffs = []
  outcome = False

  l_templ = len(str(_template))
  lp = len(_ref)
  for s in range(lp-l_templ):
    e = s + l_templ
    past_seg = str(as_numeric(_ref[s:e]))
    templ = str(_template)
    for i in range(len(past_seg)):
      diff = abs(int(past_seg[i]) - int(templ[i]))
      if diff != 0:
        diff = 1
      diffs.append(diff)
    sum_of_diffs.append(sum(diffs))
    start_idxs.append(s)
    tmp = sum(diffs)
    diffs = []
    if tmp == 0:
      outcome = True
      break
  df = pd.DataFrame({'start_idxs':start_idxs, 'sod':sum_of_diffs})

  return df, _ref, outcome
```

```python
def getSeq(str1, str2, length, LCS):
  str1Length = len(str1)
  str2Length = len(str2)
  lcs = ''
  while (str1Length > 0 and str2Length > 0):
    if str1[str1Length - 1] == str2[str2Length - 1]:
      lcs += str1[str1Length - 1]
      str1Length -= 1
      str2Length -= 1
      length -= 1
    elif (LCS[str1Length - 1][str2Length] > LCS[str1Length][str2Length - 1]):
      str1Length -= 1
    else:
      str2Length -= 1

  lcs_ = lcs[::-1]
  result = ""
  for k in range(len(lcs_)):
    if (lcs_[k] != ' '):
      result += lcs_[k]
  
  return result

def lcs(str1, str2):# longest common subsequence
  str1Length = len(str1)
  str2Length = len(str2)
  lcs = np.zeros((str1Length+1, str2Length+1))
  lcsLength = lcs.shape[0]

  for i in range(1, lcsLength):
    for j in range(1, lcs.shape[1]):
      if str1[i - 1] == str2[j - 1]:
        lcs[i][j] = 1 + lcs[i - 1][j - 1]
      else:
        lcs[i][j] = max(lcs[i - 1][j], lcs[i][j - 1])
  
  lcsString = getSeq(str1, str2, lcsLength, lcs)
  return lcsString
```
