```r
rowStats = as.data.frame(rowSds(as.matrix(gsm_form)))
row.names(rowStats) = row.names(gsm_form)
colnames(rowStats)[1] = 'sd'
rowStats$mean = rowMeans(gsm_form)
head(rowStats)
```

```r
globalStat_max = mean(rowStats$mean) + 2*sd(rowStats$sd)
globalStat_min = mean(rowStats$mean) - 2*sd(rowStats$sd)
print(mean(rowStats$mean))
print(sd(rowStats$sd))
```

```r
print(globalStat_max)
print(globalStat_min)
```

#2) Filter out genes with low expression values using some criterion.

```r
probe_keep = vector()
for (i in 1:dim(rowStats)[1]){
  stat_high = rowStats$mean[i] + 2*rowStats$sd[i]
  stat_low = rowStats$mean[i] - 2*rowStats$sd[i]
  if ((stat_low > globalStat_min) & (stat_high < globalStat_max))
    probe_keep = append(probe_keep, row.names(rowStats)[i])
}
```

```r
gsm_ = gsm_form[row.names(gsm_form) %in% probe_keep, ]
head(gsm_)
```

```r
options(repr.plot.width=10, repr.plot.height=12)
boxplot(gsm_,cex=0.9,col="lightblue",las=2,cex.axis=.8,ylab="log2(intensity)",main="After Applying Stats Threshold - Removing More Outliers")
```

#3) Conduct feature selection with a statistical test or machine learning method.

```r
t.test.all.genes <- function(x,s1,s2) {
  x1 <- x[s1]
  x2 <- x[s2]
  x1 <- as.numeric(x1)
  x2 <- as.numeric(x2)
  t.out <- t.test(x1, x2, alternative="two.sided",var.equal=T)
  out <- as.numeric(t.out$p.value)
  return(out)
}
```

```r
pv <- as.data.frame(apply(gsm_,1,t.test.all.genes,s1=C57BL_list,s2=BALB_list))
colnames(pv)[1] = 'pv'
head(pv)
```

```r
pv$probes = row.names(pv)
pv_ = as.data.frame(pv[pv$pv < 0.05, ])
```

```r
gene_filter = gsm_[row.names(gsm_) %in% row.names(pv_), ]
head(gene_filter[,1:4])
```

```r
options(repr.plot.width=10, repr.plot.height=12)
boxplot(gene_filter,cex=0.9,col="lightblue",las=2,cex.axis=.8,ylab="log2(intensity)",main="Gene Levels Post Filtering Low Expressions")
```

```r
plot(range(dat.loads[,1]),range(dat.loads[,2]),type='n',xlab='pc1',ylab='pc2',main='PCA with K-means classification')
k.class <- c(1:2)
k.col <- c("red","blue")
species <- c("C57BL/6","BALB/c")

for(i in 1:length(species)) {
  text(dat.loads[ann==species[i],1],dat.loads[ann==species[i],2],label=ann[ann==species[i]])
  for(j in 1:length(k.class)) {
    if(sum(ann==species[i] & dat.k==k.class[j])>0) {
      text(dat.loads[(ann==species[i] & 
      dat.k==k.class[j]),1],dat.loads[(ann==species[i] & 
      dat.k==k.class[j]),2],
      label=ann[(ann==species[i] & dat.k==k.class[j])],col=k.col[j])
    }
  }
}
```
